```python


-- all employees who are earning  greater than the average salary by each departments
--c : firstname, lastname, salary, department_name
--t : employees , departments
--a : avg(salary)
--g : group by department_name

with cte as (
select e.first_name, e.last_name, e.salary, d.department_name,
avg(e.salary) over (partition by e.department_id) as avgdeptsalary
from employees e
join departments d on e.department_id = d.department_id)

select * from cte
where e.salary > avgdeptsalary;



-- all employees who are earning  greater than the average salary by each departments
--c : firstname, lastname, salary, department_name
--t : employees , departments
--a : avg(salary)
--g : group by department_name

select e.first_name, e.last_name, e.salary, d.department_name,
avg(e.salary) as averagesalary from employees e
join departments d on e.department_id = d.department_id
group by e.first_name, e.last_name,e.salary, d.department_name
order by d.department_name
```

```python
--find employees who have changed departments , showing their previous and current department
--c firstname,  lastname , department_name as previousdepartment , department_name as currentdepartment
--t employee , department, jobhistory
--a none
--g none
--conditions: jh1.enddate is not null
--condition2: jh2.enddate is null
-- condition3 : jh1.department <> jh2.department_id

select e.first_name, e.last_name, d1.department_name as previousdepartment, d2.department_name as currentdepartment
from employees e
join job_history jh1 on e.employee_id = jh1.employee_id
join departments d1 on jh1.department_id = d1.department_id
join job_history jh2 on jh2.employee_id = e.employee_id
join departments d2 on jh2.department_id = d2.department_id

where jh1.end_date is not null
and jh2.end_date is NULL
and jh1.department_id <> jh2.department_id;
```

```python
#show all emoployees from a department given.

create procedure showemployeesbydepartment(in detname varchar(255))
begin
select * from employees where department = detname;
end;

call showemployeesbydepartment('HR');
```

```python
#create sp to insert a new employee to the table.

create procedure addemployee(

    in empid int,
    in empname varchar(255),
    in dept varchar(255),
    in salary int
)
begin
insert into employees (id, name, department, salary)
values (empid, empname, dept, salary);
end;

call addemployee(10, 'john', 'IT', 50000);
```
