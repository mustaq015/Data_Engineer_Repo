### Using **.withColumn()** to add a Null Columns

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import lit
from pyspark.sql.types import IntegerType

spark = SparkSession.builder.appName("DataFrame with Null Values").getOrCreate()

data = [("Mustaq", 31, 10, 30000),
        ("Ali", 32, 20, 40000),
        ("Ahmed", None, 30, 50000),
        ("Adnan", None, None, 60000),
        ("Null", 27, 23, 38000),
        ("Null", 36, None, None)]

columns = ["Employee Name", "Age of Employee", "Experience (in Years)", "Salary (per Month - $)"]

df = spark.createDataFrame(data, columns)

df_with_null = df.withColumn("value", lit(None))\
                .withColumn("Sales", lit(None).cast(IntegerType()))

display(df_with_null)
```
