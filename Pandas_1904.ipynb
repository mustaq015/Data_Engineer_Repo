```python
#this is my starting point for the project
#load the csv file
import pandas as pd
#path= "D:\Users\krish\Downloads\financedata.csv"
df = pd.read_csv('/content/data.csv')
```

```python
df.info()
```

```python
#drop Duplicates
df.drop_duplicates(inplace=True)
df.info()
```

```python
#numerical columns fill it with mean

df['Revenue'].fillna(df['Revenue'].mean(), inplace=True)
df['Expenses'].fillna(df['Expenses'].mean(), inplace=True)
df['COGS'].fillna(df['COGS'].mean(), inplace=True)
df['Operating_Expenses'].fillna(df['Operating_Expenses'].mean(), inplace=True)
df['Depreciation'].fillna(df['Depreciation'].mean(), inplace=True)
df['Amortization'].fillna(df['Amortization'].mean(), inplace=True)
```

```python
df.info()
```

```python
df['Date'].fillna('01011900', inplace=True)
```

```python
df
```

```python
df['Department'].fillna('unknown', inplace=True)
```

```python
df.info()
```

```python
#aggregrated columns
#total revenue: revenue
#total expense: expense
#profit: revenue - expense
#gross margin: (revenue - COGS)/ revenue
#operating income : revenue-operating expense
#EBITDA (earnings befre interest, tax, depreciation and amortization): operatingincome+depreciation + amortization
#accounts receivables: AR
#accounts payables. AP
#year ove year growth : 5% . yoy = revenue * 0.05


df['total_revenue'] = df['Revenue']
df['total_expense'] = df['Expenses']
df['profit'] = df['Revenue'] - df['Expenses']
df['gross_margin'] = (df['Revenue'] - df['COGS'])/df['Revenue']
df['operating_income'] = df['Revenue'] - df['Operating_Expenses']
df['EBITDA'] = df['operating_income'] + df['Depreciation'] + df['Amortization']
df['accounts_receivables'] = df['AR']
df['accounts_payables'] = df['AP']
df['year_over_year_growth'] = df['Revenue'] * 0.05
```

```python
#merge 10 files into a single file using pandas

import pandas as pd
import glob

#glob is used to list all files in a specific folder.

files = glob.glob('path/*.csv')

#use pd.concat which will merge all dataframes into one. I used a list comprehension which will read csvfiles in the fileslist and pd.concat will merge them one by one.
#ignore_index = True : reset the index based on the newdataframe

mergedf = pd.concat([pd.read_csv(file) for file in files], ignore_index=True)

#to_csv: write the new mergeddf to a new csv file

mergedf.to_csv('path', index= False)

```
