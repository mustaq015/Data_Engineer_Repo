**DataFrame** : 2 Dimensional(rows & columns)

```python
import pandas as pd
data = {
    "calories": [420, 380, 390],
    "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index=['a', 'b', 'c'])
df
```

###loc vs iloc
**loc** : Access row & column by label, used when you want to access row & column by their label name

**iloc** : Access data based on their index position.

```python
#df.loc['a', 'duration'] # i can use the name of the columns(labels) to get the data back(label-based indexing).
#df.loc['a':'b', 'duration']
df.loc[df['duration'] != 50, :]
#df.loc[df['duration'] != 50, 'duration']
```

```python
df.iloc[1, 1]   # i can only use the indices of both row & columns to get the data back(position-based indexing).
```
